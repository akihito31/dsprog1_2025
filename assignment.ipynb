## 最終課題

```python
import requests
from bs4 import BeautifulSoup
import time

starturl = "https://www.musashino-u.ac.jp/"
visited_links = set()  # 訪れたリンクを記録する
urls = {}  # URLとそのタイトルを格納する辞書

# リンクを再帰的にクロールする関数
def get_links(url):
    try:
        # 指定されたURLにリクエストを送信
        res = requests.get(url)
        res.raise_for_status()  # HTTPエラーが発生した場合は例外をスロー

        # レスポンスのHTMLコンテンツをBeautifulSoupでパース
        html_soup = BeautifulSoup(res.content, 'html.parser')

        # ページ内のすべての<a>タグを取得
        a_tags = html_soup.find_all('a')
        for a_tag in a_tags:
            href = a_tag.get('href')  # <a>タグのhref属性を取得
            if href:
                # 内部リンクかどうかを判定
                if href.startswith('https://www.musashino-u.ac.jp/') or href.startswith('/'):
                    # 相対リンクを絶対URLに変換
                    if href.startswith('/'):
                        href = starturl.rstrip('/') + href  # 相対URLを完全なURLに変換

                    # 訪れたリンクでない場合
                    if href not in visited_links:
                        visited_links.add(href)

                        # リンクのタイトルを取得
                        try:
                            link_res = requests.get(href)
                            link_soup = BeautifulSoup(link_res.content, 'html.parser')
                            title = link_soup.title.string.strip() if link_soup.title else 'No title'

                            # 辞書に保存
                            if href not in urls:
                                urls[href] = title
                                print(f'{href}: {title}')
                        except Exception as e:
                            print(f'Error accessing {href}: {e}')

                        # サーバーへの負担を軽減するため待機
                        time.sleep(0.3)

                        # 再帰的にクロール
                        get_links(href)
    except Exception as e:
        print(f'Error accessing {url}: {e}')

# クロール開始
get_links(starturl)

# 収集結果の出力
print('Links and titles:')
for idx, (link, title) in enumerate(urls.items(), start=1):
    print(f'{idx}. {link}: {title}')

# ファイル保存
with open('links_and_titles.txt', 'w', encoding='utf-8') as f:
    for link, title in urls.items():
        f.write(f'{link}\t{title}\n')
```
